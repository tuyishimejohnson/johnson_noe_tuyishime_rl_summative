```python
# Import neccessary libraries

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Simulated training data
episodes = np.arange(1, 20001, 100)
dqn_rewards = np.cumsum(np.random.normal(5, 2, len(episodes)))
ppo_rewards = np.cumsum(np.random.normal(6, 1.5, len(episodes)))

dqn_loss = np.abs(np.random.normal(0.5, 0.2, len(episodes))) 
ppo_entropy = np.abs(np.random.normal(0.6, 0.1, len(episodes)))
```

```python
# Plot Cumulative Rewards
plt.figure(figsize=(12, 5))
plt.plot(episodes, dqn_rewards, label="DQN Cumulative Reward", color='blue')
plt.plot(episodes, ppo_rewards, label="PPO Cumulative Reward", color='green')
plt.xlabel("Episodes")
plt.ylabel("Cumulative Reward")
plt.title("Cumulative Reward Over Episodes")
plt.legend()
plt.grid(True)
plt.show()

# Plot Training Stability (Loss for DQN, Policy Entropy for PPO)
plt.figure(figsize=(12, 5))
plt.plot(episodes, dqn_loss, label="DQN Loss Curve", color='violet')
plt.plot(episodes, ppo_entropy, label="PPO Policy Entropy", color='gray')
plt.xlabel("Episodes")
plt.ylabel("Loss / Entropy")
plt.title("Training Stability: DQN Loss vs. PPO Entropy")
plt.legend()
plt.grid(True)
plt.show()
```
